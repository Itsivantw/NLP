```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.ptt.cc/bbs/index.html"
resp = requests.get(url)
web_content = resp.text
#resp.encoding = 'utf-8'
#print(resp.text)
soup = BeautifulSoup(web_content, 'lxml')
arc_html = soup.find_all("div", class_ = "b-ent")
a=1
for x in arc_html:
    print(a)
    print(x)
    a+=1
```

```python
cat_href = []
for x in arc_html:
    Cate = x.find("a")
    href = Cate.get("href")
    cat_href.append(href)
    print(href)
```

```python
cat_href
```

```python
import random
import time
all_arcHref = []
de_c = [1,2,3,4,5,6]
for h in cat_href:
    #d = random.choice(de_c)
    #time.sleep(d)
    url = "https://www.ptt.cc" + h
    resp = requests.get(url)
    web_content = resp.text
    #resp.encoding = 'utf-8'
    #print(resp.text)
    soup = BeautifulSoup(web_content, 'lxml')
    title_html = soup.find_all("div", class_ = "title")
    for T in title_html:
        print(T)
        arc = T.find("a")
        if(arc != None):
            a_href = arc.get("href")
            print(a_href)
            all_arcHref.append(a_href)
    print("FININ"+ h)
```

```python
all_arcHref
```

```python
author_list =[]
board_list = []
title_list = []
date_list = []
content_lsit = []
import bs4
for arc_num in all_arcHref:
    URL = "https://www.ptt.cc" + arc_num
    # 設定Header與Cookie
    my_headers = {'cookie': 'over18=1;'}
    # 發送get 請求 到 ptt 八卦版
    response = requests.get(URL, headers = my_headers)


    #  把網頁程式碼(HTML) 丟入 bs4模組分析
    soup = bs4.BeautifulSoup(response.text,"html.parser")

    ## PTT 上方4個欄位
    header = soup.find_all('span','article-meta-value')
    print(len(header))
    if(len(header)>=4):
        # 作者
        author = header[0].text
        # 看版
        board = header[1].text
        # 標題
        title = header[2].text
        # 日期
        date = header[3].text


        ## 查找所有html 元素 抓出內容
        main_container = soup.find(id='main-container')
        # 把所有文字都抓出來
        all_text = main_container.text
        # 把整個內容切割透過 "-- " 切割成2個陣列
        pre_text = all_text.split('--\n※')[0]
    
        # 把每段文字 根據 '\n' 切開
        texts = pre_text.split('\n')
        # 如果你爬多篇你會發現 
        contents = texts[2:]
        # 內容
        content = '\n'.join(contents)


        # 顯示
        print('作者：'+author)
        author_list.append(author)
        print('看板：'+board)
        board_list.append(board)
        print('標題：'+title)
        title_list.append(title)
        print('日期：'+date)
        date_list.append(date)
        content = content.replace("★" , "").replace("\r\n" , "").replace("！","").replace("《","").replace("》","").replace("，","").replace("/","").replace("（","").replace("）","").replace("\n","").replace("-","").replace("※","")
        print('內容：'+content)
        content_lsit.append(content)
```

```python
import pandas as pd
df_ptt_id = pd.DataFrame()
df_ptt_id["作者"]=author_list
df_ptt_id["看板"]=board_list
df_ptt_id["標題"]=title_list
df_ptt_id["日期"]=date_list
df_ptt_id["內容"]=content_lsit
#存到檔案做備用
df_ptt_id.to_csv("./ptt.csv",encoding="utf-8-sig")
```

```python
from sklearn.model_selection import train_test_split
data = df_ptt_id
data.head()
```

```python
train_data , test_data = train_test_split(data, random_state = 42,test_size=0.2)
```

```python
import matplotlib.pyplot as plt

plt.figure( figsize = (20,12))
plt.subplot(1,2,1)
train_data["看板"].value_counts().sort_index().plot( kind = "barh", title = "內容")

plt.subplot(1,2,2)
test_data["看板"].value_counts().sort_index().plot( kind = "barh", title = "內容")
```

```python
import jieba
import re

def content_cut(text):
    specialChars = "!#$%^&*()《  》"
    text = text.replace("★" , "").replace("\r\n" , "").replace("！","").replace("《","").replace("》","").replace("，","").replace("/","").replace("（","").replace("）","").replace(specialChars,"")
    return " ".join(list(jieba.cut(text)))

train_data['分詞內容'] = train_data['內容'].map(content_cut)
test_data['分詞內容'] = test_data['內容'].map(content_cut)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words=None, token_pattern="(?u)\\b\\w+\\b")
X_train = vectorizer.fit_transform(train_data["分詞內容"])
X_test = vectorizer.transform(test_data["分詞內容"])
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=51,weights="distance")
knn.fit(X_train, train_data["看板"])
```

```python
Y_test = knn.predict(X_test)
```

```python
socre=knn.score(X_test,test_data["看板"])
print('精準度:{:2f}'.format(socre))
```
