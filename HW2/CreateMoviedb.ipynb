```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
# 搜尋目前上映那些電影，擷取出其ID資訊
title_list = []
id_list = []
type_list = []
story_list = []
date_list = []
for i in range(12000):
    url = "https://movies.yahoo.com.tw/movieinfo_main/" + str(i)
    resp = requests.get(url)
    web_content = resp.text
    #resp.encoding = 'utf-8'
    #print(resp.text)
    soup = BeautifulSoup(web_content, 'lxml')
    movie_html = soup.find("h1")

    print(i)
    if(movie_html != None):
        print("電影名稱:"+movie_html.string)
        title_list.append(movie_html.string)
        id_list.append(i)
        movie_typeX = soup.find("div", class_ = "level_name_box")
        movie_type = movie_typeX.find_all('a', class_ = "gabtn")
        print("分類:")
        type1 = []
        for movie_t in movie_type:
            type1.append(movie_t.string.strip('\n').strip(' ').strip('\n'))
            print(movie_t.string.strip('\n').strip(' ').strip('\n'))
        type_list.append(type1)
        movie_storyX =  soup.find('div', class_ = 'gray_infobox_inner').span.text.strip()
        #movie_story = movie_storyX.find('span')
        story_list.append(movie_storyX)
        print(movie_storyX)
        movie_dateX =  soup.find('div', class_ = 'movie_intro_info_r')
        movie_date = movie_dateX.find('span')
        print(movie_date.string)
        date_list.append(movie_date.string)
        df_movie_id = pd.DataFrame()
df_movie_id["電影"]=title_list
df_movie_id["ID"]=id_list
df_movie_id["分類"]=type_list
df_movie_id["劇情介紹"]=story_list
df_movie_id["上映日期"]=date_list
#存到檔案做備用
df_movie_id.to_csv("./movie.csv",encoding="utf-8-sig")
```
